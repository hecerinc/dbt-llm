```python
import numpy as np
import pandas as pd
```

```python
convo_gen = pd.read_pickle('./convo_generation/output/20240306_convos_full.pickle')
convo_gen
```

```python
convo_gen_df = pd.DataFrame.from_records(convo_gen).T
convo_gen_df = convo_gen_df.set_index('id')
convo_gen_df
```

```python
thread_recs = []
dbt_recs = []
patient_recs = []

for id in convo_gen_df.index:
    row = convo_gen_df.loc[id]

    thread_series = pd.Series(row['thread_stats'], name=id).add_prefix('total_')
    dbt_series = pd.Series(row['result']['dbt'], name=id).add_prefix('dbt_')
    patient_series = pd.Series(row['result']['persona'], name=id).add_prefix('patient_')

    thread_recs.append(thread_series)
    dbt_recs.append(dbt_series)
    patient_recs.append(patient_series)

thread_df = pd.DataFrame(thread_recs)
dbt_df = pd.DataFrame(dbt_recs)
patient_df = pd.DataFrame(patient_recs)

stats_df = pd.concat([patient_df, dbt_df, thread_df], axis=1).convert_dtypes()
convo_gen_df = convo_gen_df.join(stats_df)
convo_gen_df
```

```python
convo_gen_df = convo_gen_df.drop(
    ['result', 'prompt', 'thread_stats'],
    axis=1,
).rename(
    {'elapsed_time': 'total_elapsed_time'},
    axis=1,
)
convo_gen_df
```

```python
tot_time = convo_gen_df.pop('total_elapsed_time')
convo_gen_df.insert(
    len(convo_gen_df.columns),
    'total_elapsed_time',
    tot_time
)
convo_gen_df
```
